# Exploratory Data Analysis (EDA)

### Setup

```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option('display.max_columns', None)

# Load dataset
df = pd.read_csv('../data/raw/data.csv')  

# Quick look
df.head()
```

### 1. Dataset Overview

```python
# Number of rows and columns
print("Shape of dataset:", df.shape)

# Data types
print(df.dtypes)

# Summary of missing values
print(df.isnull().sum())
```

### 2. Summary Statistics

```python
# Numerical columns
df.describe().T

# Categorical columns
df.describe(include='object').T
```

### 3. Numerical Feature Distributions

```python
numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

### 4. Categorical Feature Distributions

```python

# Select only object-type columns
categorical_cols = df.select_dtypes(include='object').columns

# Columns to skip (IDs / timestamps)
skip_cols = ['TransactionId', 'BatchId', 'TransactionStartTime']

# Number of top categories to show for high-cardinality columns
top_n = 10  # smaller for assessment clarity

for col in categorical_cols:
    if col in skip_cols:
        continue  # skip non-informative columns

    counts = df[col].value_counts()
    
    # Plot only top N if there are too many unique values
    if len(counts) > top_n:
        counts = counts.nlargest(top_n)
    
    plt.figure(figsize=(6,4))
    sns.barplot(x=counts.values, y=counts.index, palette='viridis')
    plt.title(f'Top {top_n} categories for {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
    plt.close()
```

### 5. Correlation Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = df.select_dtypes(include=['int64', 'float64'])  # only numeric columns

plt.figure(figsize=(10,8))
sns.heatmap(numeric_cols.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

### 6. Missing Values

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing
```

### 7. Outlier Detection

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```
